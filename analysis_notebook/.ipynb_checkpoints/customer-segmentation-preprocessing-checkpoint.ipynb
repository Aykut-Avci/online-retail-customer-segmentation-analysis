# Customer Segmentation Preprocessing 

This notebook is used for conversion of excel files into csv and merging two datasets into one for faster data loading. These tasks are run only once to optimize data access efficiency.

## Loading Libraries

```python
import pandas as pd
```

```python
#Reading excel files
df1 = pd.read_excel(r"C:\Users\aykut\data_analytics_projects\customer_segmentation_analysis\data\online_retail.xlsx")
df2 = pd.read_excel(r"C:\Users\aykut\data_analytics_projects\customer_segmentation_analysis\data\online_retail_II.xlsx")
#Standardizing column names 
df2.columns = ['InvoiceNo', 'StockCode', 'Description', 'Quantity', 'InvoiceDate',
       'UnitPrice', 'CustomerID', 'Country']
#Merging dataframes
df_merged = pd.concat([df1,df2],axis=0,ignore_index=True)
#Saving merged csv file
df_merged.to_csv('online_retail_merged.csv',index=False)
```
